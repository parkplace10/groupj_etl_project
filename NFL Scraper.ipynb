```python
import pandas as pd
```

```python
season = 2018
week = 16
```

```python
dfs = []
pages = 1
while pages < 326:
    url = f"http://fantasy.nfl.com/research/projections?offset={pages}&position=O&sort=projectedPts&statCategory=projectedStats&statSeason={season}&statType=weekProjectedStats&statWeek={week} "
    df = pd.read_html(url,header=1)[0]
    dfs.append(df)
    pages += 25
nfl_df = pd.concat(dfs)
nfl_df.to_csv(f'NFL-week{week}-projections.csv')
```
